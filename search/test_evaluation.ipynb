```python
import json

with open("/Users/katharinazeilnhofer/VS Code/ir-project/data_retrieval/kochwiki_data/rezepte_parsed.json", "r") as f:
    recipes = json.load(f)

def find_recipes_with_content(keyword, n=10):
    results = []
    for r in recipes:
        if not r.get("zutaten_namen") or not r.get("zubereitung_raw"):
            continue
        if keyword.lower() in r["title"].lower():
            results.append(r["title"])
    return results[:n]

queries = ["Kartoffelsuppe", "Gulasch", "Schokoladentorte", "Weihnachtsplätzchen", 
           "Sauerteig", "japanisch", "Curry", "Nudel", "Soße", "Kartoffel"]

for q in queries:
    results = find_recipes_with_content(q)
    print(f"\n'{q}' ({len(results)} mit Inhalt):")
    for r in results[:7]:
        print(f"  {r}")
```
