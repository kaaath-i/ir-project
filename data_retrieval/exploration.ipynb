```python
import json

with open("kochwiki_data/rezepte_parsed.json", "r") as f:
    recipes = json.load(f)

with open("kochwiki_data/zutaten_parsed.json", "r") as f:
    zutaten = json.load(f)

print(f"Recipes: {len(recipes)}")
print(f"Ingredients: {len(zutaten)}")

print("\nFirst 10 Recipes:")
for r in recipes[:10]:
    print(f"  - {r['title']}")

print(f"\nFirst 10 Ingredients:")
for z in zutaten[:10]:
    print(f"  - {z['name']}")
```

```python
cuisines = {}
for r in recipes:
    for k in r.get("kueche", []):
        cuisines[k] = cuisines.get(k, 0) + 1

print("Top 15 Cuisines:")
for k, v in sorted(cuisines.items(), key=lambda x: -x[1])[:15]:
    print(f"  {v:>4}x  {k}")


print("\nLevels:")
diff = {}
for r in recipes:
    d = r["metadata"]["schwierigkeit"] or "unbekannt"
    diff[d] = diff.get(d, 0) + 1
for k, v in sorted(diff.items(), key=lambda x: -x[1]):
    print(f"  {v:>4}x  {k}")

print("\nRecipes with the most ingredients:")
by_zutaten = sorted(recipes, key=lambda r: len(r["zutaten"]), reverse=True)
for r in by_zutaten[:5]:
    print(f"  {len(r['zutaten']):>3} Ingredients: {r['title']}")
```

```python
def simple_search(query, data=recipes, field="plaintext", limit=5):
    query = query.lower().split()
    results = []
    for item in data:
        text = item.get(field, "").lower()
        matches = sum(1 for q in query if q in text)
        if matches > 0:
            results.append((matches, item["title"]))
    results.sort(reverse=True)
    return results[:limit]
```

```python
queries = ["Kartoffelsuppe",
    "Dessert vegan",
    "Weihnachtsplätzchen",
    "schnelles Mittagessen Nudeln",
    "französische Soße",
    "Brot backen Sauerteig",
    "Gulasch",
    "Kuchen glutenfrei",
    "japanische Suppe",
    "Schokoladentorte"]

for q in queries:
    print(f"\nSearch for '{q}':")
    results = simple_search(q)
    print(f"\n'{q}':")
    for score, title in results[:7]:
        print(f"   [{score}] {title}")
```

```python
searches = [
    ("Kartoffelsuppe", "title"),
    ("Gulasch", "title"),
    ("Schokoladentorte", "title"),
    ("Plätzchen", "title"),
    ("Sauerteig", "title"),
    ("vegan Dessert", "title"),
    ("glutenfrei", "title"),
    ("japanisch", "title"),
    ("Nudel", "title"),
    ("Soße französisch", "title"),
]

for q, field in searches:
    results = simple_search(q, field=field, limit=7)
    print(f"\n'{q}' (in {field}):")
    for score, title in results:
        print(f"   [{score}] {title}")
```

```python
extra = [
    ("vegan", "title"),
    ("Dessert", "title"),
    ("glutenfrei", "plaintext"),
    ("Soße", "title"),
    ("französisch", "plaintext"),
    ("Ramen", "title"),
    ("Miso", "title"),
    ("Weihnacht", "title"),
    ("Brot backen", "title"),
    ("Curry", "title"),
]

for q, field in extra:
    results = simple_search(q, field=field, limit=7)
    print(f"\n'{q}' (in {field}):")
    for score, title in results:
        print(f"   {title}")
```

```python
evaluation = {
    "Kartoffelsuppe": ["Vegetarische Kartoffelsuppe", "Vegane Kartoffelsuppe", "Leipziger Kartoffelsuppe", "Leas Kartoffelsuppe", "Lauch-Kartoffelsuppe"],
    "Gulasch": ["Würstchengulasch mit Paprika", "Wurstgulasch", "Winterlicher Krautgulasch", "Wildschweingulasch mit Pilzen und Senfsauce", "Wildschweingulasch mit Pilzen"],
    "Schokoladentorte": ["Himmlische Schokoladentorte", "Schokoladentorte, himmlisch", "Schokoladentorte mit Mascarpone-Minze-Creme", "Maroni-Schokoladentorte", "Brigittes Schokoladentorte"],
    "Weihnachtsplätzchen": ["Walisische Weihnachtsplätzchen", "Waliser Weihnachtsplätzchen", "Detmolder Weihnachtsplätzchen", "Zimtsterne mit Mandeln", "Plätzchen zum Ausstechen"],
    "Brot mit Sauerteig": ["Sauerteigbrot ohne Hefe aus dem Backautomaten", "Sauerteigbrot (hefefrei) aus dem Backautomaten", "Römisches Sauerteigbrot", "Roggenbrot auf Sauerteigbasis", "Kartoffelbrot mit Sauerteig"],
    "japanische Suppe": ["Misosuppe mit Tofu", "Miso-Suppe mit Tofu", "Ramen (Japanische Nudelsuppe)", "Japanische Nudelsuppe", "Miso Shiru"],
    "Curry": ["Tomaten-Linsen-Curry", "Thailändisches Gemüsecurry", "Veganes Blumenkohlcurry", "Thai-Lychee-Curry", "Zanderfilet mit Apfel-Gemüse-Curry"],
    "schnelles Mittagessen mit Nudeln": ["Zitronennudeln", "Studenten-Nudelauflauf", "Tagliatelle in Schinken-Sahnesauce", "Tagliatelle mit Lachs-Sherrysauce", "Zha-Jiang Nudeln nach Peking-Art"],
    "französische Soße": ["Sauce ravigote", "Vinaigrette mit roter Paprikaschote", "Vinaigrette mit Feta", "Erdbeervinaigrette", "Walnusssoße"],
    "glutenfreier Kuchen": ["Buchweizenpfannkuchen", "Hirseauflauf mit Kirschen", "Spanischer Orangenpudding", "Heidensterz", "Geröstete Maronen"],
}

print("EVALUATION:\n")
total_hits = 0
total_possible = 0

for query, expected in evaluation.items():
    results = simple_search(query, field="plaintext", limit=5)
    retrieved_titles = [title for _, title in results]
    
    hits = sum(1 for doc in expected if doc in retrieved_titles)
    total_hits += hits
    total_possible += len(expected)
    
    print(f"Query: '{query}'")
    print(f"  Hits: {hits}/5")
    print(f"  Retrieved: {retrieved_titles}")
    print(f"  Expected:  {expected}")
    print()

print(f"{'='*50}")
print(f"Total: {total_hits}/{total_possible} relevant documents found")
print(f"Recall@5: {total_hits/total_possible:.1%}")
```
